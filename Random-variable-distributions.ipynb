Compruebe mediante un gráfico de distribución si las variables se ajustan a la distribución teórica normal o exponencial.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Suponiendo que tienes un DataFrame llamado `numerical_df` con solo datos numéricos
for column in numerical_df.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(numerical_df[column], kde=True, color='blue')
    plt.title(f'Histograma de {column}')
    plt.xlabel(column)
    plt.ylabel('Frecuencia')
    plt.show()
```

Gráficos Q-Q para una distribución normal

```python
for column in numerical_df.columns:
    plt.figure(figsize=(6, 4))
    stats.probplot(numerical_df[column].dropna(), dist="norm", plot=plt)
    plt.title(f'Gráfico Q-Q para una distribución normal de {column}')
    plt.show()
```

Gráficos Q-Q para una distribución exponencial:

```python
for column in numerical_df.columns:
    plt.figure(figsize=(6, 4))
    stats.probplot(numerical_df[column].dropna(), dist="expon", plot=plt)
    plt.title(f'Gráfico Q-Q para una distribución exponencial de {column}')
    plt.show()
```

Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

Vamos a aplicar una transformación logarítmica y visualizar los resultados.

```python
# Asegúrate de que los datos son positivos antes de aplicar log transform
numerical_df = numerical_df[numerical_df > 0]  # Asumiendo que tus datos pueden ser cero o negativos en algunos casos

# Aplicar la transformación logarítmica
numerical_df_log = np.log(numerical_df)

# Mostrar histogramas para datos transformados
for column in numerical_df_log.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(numerical_df_log[column].dropna(), kde=True, color='blue')  # Uso dropna() para evitar problemas con log(0)
    plt.title(f'Log-transformed Histogram of {column}')
    plt.xlabel(f'Log of {column}')
    plt.ylabel('Frequency')
    plt.show()

    # Crear gráfico Q-Q para la distribución normal
    plt.figure(figsize=(6, 4))
    stats.probplot(numerical_df_log[column].dropna(), dist="norm", plot=plt)
    plt.title(f'Q-Q Plot for Log-transformed {column}')
    plt.show()
```
